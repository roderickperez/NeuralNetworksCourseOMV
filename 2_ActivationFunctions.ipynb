```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Enable non-interactive mode for Jupyter
%matplotlib inline

# Define the rectified function (ReLU)
def rectified(x):
    return max(0, x)

# Define input data
inputs = [x for x in range(-10, 10)]
outputs = [rectified(x) for x in inputs]

# Create a slider widget
slider = widgets.IntSlider(min=-10, max=9, step=1, value=0, description='x value')

# Function to update the plot based on slider value
def update_plot(change):
    x_val = slider.value
    y_val = rectified(x_val)
    
    # Clear the previous output (including the slider) and re-display
    clear_output(wait=True)
    
    # Display the slider again
    display(slider)
    
    # Create the plot
    plt.figure(figsize=(6, 4))
    plt.plot(inputs, outputs, label='ReLU function')
    plt.scatter([x_val], [y_val], color='red', zorder=5)
    plt.grid()
    plt.legend()
    
    # Update the title to show both x and y values
    plt.title(f"ReLU at x = {x_val}, y = {y_val}")
    
    # Show the plot
    plt.show()

# Attach the update function to slider value change
slider.observe(update_plot, names='value')

# Display the slider
display(slider)

# Initial plot
update_plot(None)
```

## Sigmoid 

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Enable non-interactive mode for Jupyter
%matplotlib inline

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define input data
inputs = np.linspace(-10, 10, 100)  # Use more points for a smoother sigmoid curve
outputs = [sigmoid(x) for x in inputs]

# Create a slider widget
slider = widgets.IntSlider(min=-10, max=10, step=1, value=0, description='x value')

# Function to update the plot based on slider value
def update_plot(change):
    x_val = slider.value
    y_val = sigmoid(x_val)
    
    # Clear the previous output (including the slider) and re-display
    clear_output(wait=True)
    
    # Display the slider again
    display(slider)
    
    # Create the plot
    plt.figure(figsize=(6, 4))
    plt.plot(inputs, outputs, label='Sigmoid function')
    plt.scatter([x_val], [y_val], color='red', zorder=5)
    plt.grid()
    plt.legend()
    
    # Update the title to show both x and y values
    plt.title(f"Sigmoid at x = {x_val}, y = {y_val:.4f}")
    
    # Show the plot
    plt.show()

# Attach the update function to slider value change
slider.observe(update_plot, names='value')

# Display the slider
display(slider)

# Initial plot
update_plot(None)
```

## Tanh

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Enable non-interactive mode for Jupyter
%matplotlib inline

# Define the tanh function
def tanh(x):
    return np.tanh(x)

# Define input data
inputs = np.linspace(-10, 10, 100)  # More points for a smoother tanh curve
outputs = [tanh(x) for x in inputs]

# Create a slider widget
slider = widgets.IntSlider(min=-10, max=10, step=1, value=0, description='x value')

# Function to update the plot based on slider value
def update_plot(change):
    x_val = slider.value
    y_val = tanh(x_val)
    
    # Clear the previous output (including the slider) and re-display
    clear_output(wait=True)
    
    # Display the slider again
    display(slider)
    
    # Create the plot
    plt.figure(figsize=(6, 4))
    plt.plot(inputs, outputs, label='tanh function')
    plt.scatter([x_val], [y_val], color='red', zorder=5)
    plt.grid()
    plt.legend()
    
    # Update the title to show both x and y values
    plt.title(f"tanh at x = {x_val}, y = {y_val:.4f}")
    
    # Show the plot
    plt.show()

# Attach the update function to slider value change
slider.observe(update_plot, names='value')

# Display the slider
display(slider)

# Initial plot
update_plot(None)
```
